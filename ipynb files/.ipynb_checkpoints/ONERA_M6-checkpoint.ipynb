Process data identically to Constantine

```python
df = pn.DataFrame.from_csv('/Users/7ag/active-manifolds/code_anthony/as-data-sets/ONERA-M6/ONERA-M6.txt')
data = df.as_matrix()

#Unnoramlized input values
X = data[:,:50]

#Output values
F_lift = data[:,50]
F_drag = data[:,51]

#gradient values
G_lift = data[:,52:102]
G_drag = data[:,102:]

normGlift = preprocessing.normalize(G_lift)
normGdrag = preprocessing.normalize(G_drag)

M, m = X.shape
labels = df.keys()
out_labels = labels[50:52]

xl = -0.05*np.ones((1, m))
xu = 0.05*np.ones((1, m))
XX = BoundedNormalizer(xl, xu).normalize(X)
```

## Lift experiment

```python
np.random.seed(43)
am, fVals = build_AM_from_data(np.ravel(2*np.random.rand(50,1)-1), XX, F_lift, normGlift, 0.03)

splinePlot(am,fVals, 'C_L','L')

numpts = len(am)
sValues = np.linspace(0., numpts, numpts) / (numpts)

fig=plt.figure()
ax = fig.add_subplot(111)
ax.set_title(r'Coordinate Derivatives Along $\gamma_{L}(t)$', fontsize = 13)
ax.set_xlabel(r'Curve Parameter: $t$', fontsize = 13)
[ax.plot(sValues,np.abs(np.gradient(am[:,i]))) for i in xrange(50)]
plt.xlim(0.1,0.5)
plt.savefig('OneraL_derivs.pdf')
```

### Lift errors

```python
mainRandEx(43, 50, XX, F_lift, normGlift, 0.03, 247, 'OneraL-AM')
SubspEx(43, XX, F_lift, G_lift, 247, 'OneraL-AS')
```
