```python
from functions.plotfuncs import *
from functions.ground.base_am_fxns import *
from functions.error_mains import *

#Need these for some plots
from math import pi
def r(x,y): return np.sqrt(x**2+y**2)
def f(x,y): return x**2+3*y**2
def g(x,t): return t*x**3
def h(x,y): return x**2+y**2 - np.log(x**2)
def l(x,t): return t*x
def c(x,y): return y/x

x_ = np.linspace(-1,1,num=1000)
y_ = np.linspace(-1,1,num=1000)
x,y = np.meshgrid(x_,y_)
xplus = filter(lambda x: x>0, x_)
```

## Sum of Squares

```python
# Get inputs and outputs
meshy = make_mesh(0.01,2)
sqGrads = compute_paths(gradSquaresum,meshy)[0]
squareSamps = sample_f_on_mesh(Squaresum,meshy)

# Build AM (starting away from origin for best results)
am, fVals = build_AM_from_data([0.5,0.5], meshy, squareSamps, sqGrads, 0.005)

# Fit the data with cubic splines
splinefitSS(am,fVals)

# Quantities for plots
numpts = len(am)
sValues = np.linspace(0., numpts, numpts) / (numpts)

# Coordinate Plot
fig=plt.figure()
ax = fig.add_subplot(111)
ax.set_title(r'Coordinate values of $\gamma(t)$', fontsize = 13)
ax.set_xlabel(r'Curve Parameter: $t$', fontsize = 13)
plt.plot(sValues,am[:,0], c='#fc8d62', label = "x(t)", markersize=3, linewidth=2)
plt.plot(sValues,am[:,1], c='#66c2a5', label = "y(t)", markersize=3, linewidth=2)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('Squaresum_coords.pdf', bbox_inches = 'tight')

# Derivative Plot
fig=plt.figure()
ax = fig.add_subplot(111)
ax.set_title(r'Coordinate Derivatives Along $\gamma(t)$', fontsize = 13)
ax.set_xlabel(r'Curve Parameter: $t$', fontsize = 13)
plt.plot(sValues,np.abs(np.gradient(am[:,0])), c='#fc8d62', label = "|x'(t)|", markersize=2, linewidth=2)
plt.plot(sValues,np.abs(np.gradient(am[:,1])), c='#66c2a5', label = "|y'(t)|", markersize=2, linewidth=2)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim(0,1)
plt.ylim(0.003,0.004)
plt.savefig('Squaresum_derivs.pdf',bbox_inches = 'tight')

# Orthogonal Family Plot
fig=plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Circles and Orthogonal Family', fontsize = 13)
ax.set_xlabel(r'$x$', fontsize = 13)
ax.set_ylabel(r'$y$', fontsize = 13)
ls=plt.contour(x,y, r(x,y),10, colors = '#b3b3b3')
plt.xlim(-1,1)
plt.ylim(-1,1)
[plt.plot(x_,l(x_,t), color = cm.viridis(3*t), zorder=10) for j in range(-3,4) for t in [2**j]]
[plt.plot(x_,-l(x_,t), color = cm.viridis(3*t), zorder=10) for j in range(-3,4) for t in [2**j]]
plt.plot(np.zeros(len(y)),y_, color = cm.viridis(0), zorder=10)
plt.plot(x_,np.zeros(len(x)), color = cm.viridis(100), zorder=10)
plt.plot(xplus, np.array(xplus), linewidth=5, color = '#401A4C',zorder=10)
plt.savefig('circfam.pdf')
```

## Chris' functions

```python
meshy = make_mesh(0.02,2)

def f1(x,y): return np.exp(y-x**2)
def gradf1(x,y): return [-2*x*np.exp(y-x**2), np.exp(y-x**2)]

f1Samps = sample_f_on_mesh(f1, meshy)
gradsf1 = compute_paths(gradf1,meshy)[0]
asgradsf1 = map(lambda x: gradf1(*x), meshy)

mainRandEx(42,2,np.array(meshy),f1Samps,np.array(gradsf1),0.01,8000,'f1-AM')
SubspEx(42,np.array(meshy),f1Samps,np.array(asgradsf1),8000,'f1-AS')
```

```python
def f3(x,y): return x**3 + y**3 + 0.2*x + 0.6*y
def gradf3(x,y): return [3*x**2+0.2, 3*y**2+0.6]

f3Samps = sample_f_on_mesh(f3,meshy)
gradsf3 = compute_paths(gradf3,meshy)[0]
asgradsf3 = map(lambda x: gradf3(*x), meshy)

mainRandEx(42,2,np.array(meshy),f3Samps,np.array(gradsf3),0.01,8000,'f3-AM')
SubspEx(42,np.array(meshy),f3Samps,np.array(asgradsf3),8000,'f3-AS')
```
