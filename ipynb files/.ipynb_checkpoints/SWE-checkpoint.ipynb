# Shallow Water Equations Experiment

This notebook compares the methods of Active Subspaces and Active Manifolds for the approximation of scalar functions which depend on the solution of the parameterized multi-layer shallow water equations (SWEs).

The multi-layer SWEs in correspondence to the classic SWEs reads: for $k=1,\dots,L$,
$\newcommand{\lb}{\left (}
\newcommand{\rb}{\right )}
\newcommand{\lc}{\left \langle}
\newcommand{\rc}{\right \rangle}
\newcommand{\bu}{\mathbf u}
\newcommand{\bk}{\mathbf k}
\newcommand{\btau}{\boldsymbol \tau}
\newcommand{\bD}{\mathbf D}$

$$ \begin{equation}\label{RSW:eq}
	\left\{
	\begin{split}
    \partial_th_k&=-\nabla\cdot\lb h_k\bu_k\rb \qquad &\text{ in } \Omega, \\
        \partial_t\bu_k&=-\nabla\lb K[\bu_k]+\lb g/\rho_k\rb p_k[h]\rb-q[h_k,\bu_k]\hat{\bk}\times\lb h_k\bu_k\rb+\mathcal{G}_k\lb h,\bu\rb \qquad&\text{ in } \Omega,
	\end{split}
	\right.
	\end{equation} $$
    
where $h_k$ and $\bu_k$ are the thickness and velocity of layer $k$, $K[\bu_k]= (1/2)|\bu_k|^2$ is the kinetic energy, $\hat{\bk}\times\bu_k$ is the perpendicular velocity (often denoted $\bu_{k}^{\perp}$), and $q[h_k,\bu_k]=(1/h_k)(\hat{\bk}\cdot\nabla\times\bu_k+f)$ is the potential vorticity with Coriolis parameter $f$. 

These equations are parameterized through the force term $\mathcal{G}_k\lb h,\bu\rb$, which includes the surface wind forcing applied in the top layer, bottom drag applied in the bottom layer, and some possible (artificial) biharmonic viscosity. Briefly, these parameters are described as follows.

#### Wind forcing:  
The wind forcing is defined as
	\begin{equation}
	{\mathbf f_{\text{wind}}}=\frac{\btau_{\text{wind}}}{\rho_1 h_1}{\color{blue}\mu_1},
	\end{equation}
   
where $\rho_1$ the density of the first layer and $\btau_{\text{wind}}=\mathbf{n}\cdot{\tau_{\text{wind}}}$ is the wind profile.  Note that $\tau_{\text{wind}}$ has two components: "windStressZonal" and "windStressMeridional". In SOMA test, windStressMeridional is 0.  The parameter $\color{blue}\mu_1 \in [1,4]$ is introduced to scale the wind forcing.

#### Bottom drag:
The bottom drag is defined as
	\begin{equation}
	\mathbf f_{\text{drag}}=-c_{\text{drag}}\frac{|\bu_L|\bu_L}{\rho_Lh_L}{\color{blue}\mu_2}.
	\end{equation}
    
where $\rho_L$ is the density of layer L and $c_{\text{drag}}$ is the dimensionless bottom drag coefficient (currently set as 1e-3). The parameter $\color{blue}\mu_2 \in [1,4]$ is introduced to scale the bottom drag.

#### Biharmonic viscosity:
The biharmonic turbulence closure is
	\begin{equation}
	\mathbf D_{h}^{\bu}=-\frac{\nu_{h}}{\rho_{m}^{3/4}}\nabla^4\bu,
	\end{equation}
   
where $\rho_{m}$ is the mesh density and ranges from 0 to 1. Note that $\rho_m$ is 1 in the highest resolution region, as well as for a uniform grid.  Moreover, $\nu_{h}$ is the biharmonic coefficient which has a default value of 5.0e13 in MPAS-Ocean. Conversely, in SOMA testing, the value of $\nu_h$ varies along with the mesh resolution. In particular, its value is set to 2e10 for a 16 km resolution mesh, 2e9 for 8 km resolution, and 4e8 for 4 km resolution.

#### Initial velocity:
The initial velocity $\bu(0)={\color{blue}\mu_3}$ is parameterized using $\color{\blue}\mu_3 \in [0,5]$ as a scaling parameter.


Given this, the goal is to approximate the total kinetic energy

$$ \mathcal{K}(\mathbf{u}) = \sum_k \int_\Omega K[\bu_k]\,d\sigma, $$

where $(\mu_1, \mu_2, \mu_3) \in [1, 4] \times [1, 4] \times [0, 5]$ are sampled randomly.  By using sample values of $\mathcal{K}$ and its derivatives $\mathcal{K}_{\boldsymbol{\mu}}$ obtained from simulations of the above PDE, it is possible to train an approximation to these functions at any $\boldsymbol \mu$ in parameter space.

### Preprocessing of initial mesh data

I included random partitioning of the 500 samples to get an idea of how the sample size influences the simulation.

```python
# Import of source data
data = np.loadtxt('/Users/anthonygruber/Desktop/Projects/AMv2/data/SOMA/SOMA_500_cases.dat')

# Upper and Lower bounds for parameters
UBs = np.array( [4, 4, 5] )
LBs = np.array( [1, 1, 0] )

idx = [0,0,0,0]
points = [0,0,0,0]
KVals = [0,0,0,0]
grads = [0,0,0,0]
nuPoints = [0,0,0,0]
nuGrads = [0,0,0,0]
np.random.seed(0)

for i in range(4):
  # Index sets for random partitioning
  idx[i] = np.random.randint(0, 499, 4*5**i, dtype='l')

  # Data separation
  points[i] = data[:,:3][idx[i]]
  KVals[i] = data[:,3][idx[i]]
  grads[i] = data[:,4:][idx[i]]

  # Rescaling to appropriate range
  nuPoints[i] = 2 / (UBs - LBs) * (points[i] - LBs) - 1
  nuGrads[i] = 0.5 * (UBs - LBs) * grads[i]  # Chain rule
```

## Comparison of AM and AS

Here, there is no question that AS is superior.  Even with only 20 samples and a 1-D subspace, AS produces less than 0.1% error.

This is indicative of a near-perfect linear trend in the data.

```python
# Main routine comparing AM and AS -- first on Q2. 
# Fourth input to nm.subspEx gives dimension of AS -- this can be tweaked.
for i in range(4):
  print( 'The number of samples is %i' %len(points[i]) )
  nm.manifoldEx( nuPoints[i], KVals[i], nuGrads[i],
                0.01, 0.2, nm.get_random_init_pt(3, 0), 0, False )
    
  nm.subspEx( nuPoints[i], KVals[i], nuGrads[i], 1, 0, 0.2, False, False )
  print(" ")
```

## Visualization on 100 point data set

As before, it is useful to get a picture of the results.  It is not surprising that AS behaves so well, due to the fact that the first eigenvalue is >6 orders of magnitude larger than the others.

```python
nm.manifoldEx( nuPoints[2], KVals[2], nuGrads[2],
              0.01, 0.2, nm.get_random_init_pt(3, 0), 0, True )

nm.subspEx( nuPoints[2], KVals[2], nuGrads[2], 1, 0, 0.2, True, False )
```

### Plots of coordinates and sensitivities along the AM

```python
am, fvals = nm.build_AM_from_data( nm.get_random_init_pt(3, 0),
                                  nuPoints[2], KVals[2], nuGrads[2], 0.05)

nuAM = LBs + 0.5 * (UBs - LBs) * (1 + am)

# Coordinate Plot
numpts = len(am)
sValues = np.linspace(0., numpts, numpts) / numpts
fig=plt.figure()
ax = fig.add_subplot(111)
ax.set_title(r'Coordinate Values Along $\gamma(t)$', fontsize = 13)
ax.set_xlabel(r'Curve Parameter: $t$', fontsize = 13)
plt.plot(sValues, nuAM[:,0], '-o', c='#66c2a5',
         label = r'$\mu_1$-coordinate', markersize=3, linewidth=2)
plt.plot(sValues, nuAM[:,1], '-o', c='#fc8d62',
         label = r'$\mu_2$-coordinate', markersize=3, linewidth=2)
plt.plot(sValues, nuAM[:,2], '-o', c='#8da0cb',
         label = r'$\mu_3$-coordinate', markersize=3, linewidth=2)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim(0,1)
plt.show()

# Derivative Plot
numpts = len(am)
sValues = np.linspace(0., numpts, numpts) / numpts
fig=plt.figure()
ax = fig.add_subplot(111)
ax.set_title(r'Coordinate Derivatives Along $\gamma(t)$', fontsize = 13)
ax.set_xlabel(r'Curve Parameter: $t$', fontsize = 13)
plt.plot(sValues,np.abs(np.gradient(nuAM[:,0])), '-o', c='#66c2a5',
         label = r'$|K_{\mu1}|$', markersize=3, linewidth=2)
plt.plot(sValues,np.abs(np.gradient(nuAM[:,1])), '-o', c='#fc8d62',
         label = r'$|K_{\mu2}|$', markersize=3, linewidth=2)
plt.plot(sValues,np.abs(np.gradient(nuAM[:,2])), '-o', c='#8da0cb',
         label = r'$|K_{\mu3}|$', markersize=3, linewidth=2)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim(0,1)
plt.show()
```
